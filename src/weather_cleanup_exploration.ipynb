```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Import API key
from config import meteo_key


```

```python
url = "https://api.meteostat.net/v2/point/daily"


headers = ({'x-api-key':meteo_key})
params = ({'lat': '40.757786', 
           'lon': '-73.986347', 
           'start': '2019-01-01',
           'end': '2019-06-01'           
          })

response = requests.get(url, params = params, headers = headers).json()
response

# curl --header "x-api-key: kaPAMj3qAxjQQ3xglN9MT1uNBTH2BWDr" "https://api.meteostat.net/v2/point/daily?lat=40.757786&lon=-73.986347&start=2019-01-01&end=2019-06-30"

ny_df = pd.DataFrame(response['data'])

ny_cleaned = ny_df.rename(columns = {'tavg': 'Avg Temp', 
              'tmin': 'Min Temp',
              'tmax': 'Max Temp',
              'prcp': 'Precipitation', 
              'snow': 'Snowfall', 
              'wdir': 'Wind Direction',
              'wspd': 'Wind Speed',
              'wpgt': 'Wind Peak Gust',
              'pres': 'Air Pressure',
              'tsun': 'Total Sun Duration'
             })

ny_cleaned.head()
```
